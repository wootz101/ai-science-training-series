```python
import matplotlib.pyplot as plt
import numpy as np

#Buffer size static at 8

threads = np.array([4, 5, 6, 7, 8, 9, 10])
ims = np.array([1141.31, 1168.22, 1032.09, 967.21, 705.93, 685.45, 666.23])
sd = np.array([215.24, 42.95, 73.87, 121.43, 48.29, 74.29, 58.54])


plt.figure(figsize=(6,3))
plt.plot(threads, ims)
plt.title('Threads vs Mean Images/Second')
plt.xlabel('Threads (2^x)')
plt.ylabel('Mean Images')

plt.figure(figsize=(6,3))
plt.plot(threads, sd, 'r')
plt.title('Threads vs St Dev of Images');
plt.xlabel('Threads (2^x)')
plt.ylabel('St. Dev of Images')
```
